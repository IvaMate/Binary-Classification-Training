```python
import pandas as pd
import numpy as np
import mglearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix

#Unos podataka
df = pd.read_csv('titanic_dataset_all_labeled_cleaned.csv')
X_train.head()
```

```python
#odvojeno
df_class=df[['Survived']]
df= df[['Pclass','Age','Sex_Code', 'Embarked_Code', 'Title_Code', 'AgeBin_Code',
       'FareBin_Code']]
#SPLIT
X_train, X_test, y_train, y_test = train_test_split(df, df_class, random_state=0)

print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))

print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
#VIZUALIZACIJA
#titanic_dataframe = pd.DataFrame(X_train, columns=df)
grr = scatter_matrix(X_train, c=np.ravel(y_train,order='C'), figsize=(10, 10),marker='o', hist_kwds={'bins': 20},
                        s=60, alpha=0.2,cmap=mglearn.cm3)
```

```python
#MODEL k-nearest neighbour

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, np.ravel(y_train,order='C'))
```

```python
#predictions
X_new = np.array([[2, 18, 0, 2, 2, 1, 1]])
print("X_new.shape: {}".format(X_new.shape))
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(
 df['Survived'][prediction]))
```

```python
#EVALUACIJA MODELA
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
#print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
